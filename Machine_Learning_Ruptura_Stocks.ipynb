***PRESENTACION DEL PROYECTO DE MACHINE LEARNING PARA PREDECIR RUPTURAS DE STOCK***

__Ruptura de stock__:  Una ruptura de stock ocurre cuando una empresa se queda sin existencias de un producto y no puede satisfacer la demanda de los clientes.

Las causas más habituales son:
- Previsión incorrecta de la demanda
- Retrasos de proveedores
- Problemas logísticos
- Incrementos inesperados de ventas
- Errores en la gestión de inventario
